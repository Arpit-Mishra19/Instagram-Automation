```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time
from selenium.common.exceptions import TimeoutException, NoSuchElementException

def login_instagram(driver, username, password):
    driver.get('https://www.instagram.com/accounts/login/')
    time.sleep(3)  # Wait for the page to load

    # Find the username and password fields and input your credentials
    username_field = driver.find_element(By.NAME, 'username')
    password_field = driver.find_element(By.NAME, 'password')
    
    username_field.send_keys(username)
    password_field.send_keys(password)
    password_field.send_keys(Keys.RETURN)
    
    time.sleep(5)  # Wait for the login process to complete

    print("Successfully logged in!")

def handle_pop_up(driver):
    try:
        # Wait for the pop-up to appear and then find the "Not Now" button
        not_now_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Not Now')]"))
        )
        not_now_button.click()
        print("Handled the pop-up message!")
    except TimeoutException:
        print("No pop-up found or another issue occurred")

def search_user(driver, search_query):
    try:
        # Find the search bar
        search_bar = WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.XPATH, "//input[@placeholder='Search']"))
        )
        search_bar.send_keys(search_query)
        time.sleep(3)  # Wait for search results to load
        
        # Look for the first matching user in the search results
        search_results = WebDriverWait(driver, 15).until(
            EC.presence_of_all_elements_located((By.XPATH, f"//a[contains(@href, '/{search_query}') or contains(@href, '/explore/tags/{search_query}/')]"))
        )

        if search_results:
            search_results[0].click()  # Click on the first matching result
            print(f"Navigated to the profile of {search_query}")
        else:
            print(f"No exact match found for {search_query}")
    
    except TimeoutException:
        print(f"TimeoutException: Could not find the user {search_query}")
    except NoSuchElementException:
        print(f"NoSuchElementException: Search results not found for {search_query}")

# Set up Chrome options
options = Options()
options.add_argument('--start-maximized')  # Open the browser in full screen
options.add_argument('--disable-infobars')  # Disable info bars

# Automatically handle ChromeDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# Replace with your Instagram username and password
username = 'xxxxxxxx'
password = 'xxxxxxxx'

# Replace with the Instagram username you want to search for
search_query = 'xxxxxxxx'

# Log in to Instagram
login_instagram(driver, username, password)

# Handle the pop-up if it appears
handle_pop_up(driver)

# Search for the user and navigate to their profile
search_user(driver, search_query)

# The browser will remain open, and the WebDriver session will stay active.
```
